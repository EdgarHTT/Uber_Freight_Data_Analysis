# **Customer Demands Business Case**
## **Problem statement:**
Find two locations to open the warehouses that service the provided **customer demands**. the best solution is the one that have the **minumum total distance** from the warehouses to all customer locations for **all shipments**.
## **Assumptions:**
* **Distance** between 2 locations can be calculated based on their **latitude and longitude**
* You can use any software or tools to find the answer.
* A customer can be serviced **only by one warehouse**.
* Field "Shipments" is the number of loads to be shipped from a warehouse to the customer.
* **The total distance calculation must consider the total shipments of each of the customers.**
* Rate per mile **$1.7** for each individual shipment.
* No capacity limit for each of the warehouses

## **1. Identifing the problem**
* **Context:** From a dataset of 370 different customer cities, I must find two locations of overall minimun distance for opening warehouses. Each customer can only be served by one warehouse and the warehouses have no capacity limit. The amount of shipments per customer range from 1 to 454 and they are spread all over the United States.
* **Problem to solve:** Find the two coordinates where the sum of the distance traveled is at it's minimum.
* **Why?:** To find the most efficient location to open two warehouses.

## **2. Research and Refinement** 

**Solution:** Single Source Shortest path
1. Turn the USA map into an array of coordinates
2. Each array calculates the distance to each of the customers and is multiplied by the amount of shipments
3. We select the two nodes with the smallest total distance sum that are also completely distinct (no collisions)
4. We compare each of their customers and the lowest distance keeps the customer.
5. We rerun the algorithm but now the sources do not share customers
6. We compare the coordinates in a real map and find the closest suitable place to place the warehouse

To find the best solution i decided to break down the US map into an array of coordinates, where each cell would compute the sum of all their paths to the customers and then obtaining the locations where the minimum value is found i will select two sectors. From there the sectors will compare their paths to each other and the shortest one will keep the customer. Once the customers have been reparted, the distances will be calculated again, now with their own customers, and from there we can then select the new minimum distance and rerun the calculations now with a higher a resolution.

**Tools:**
* **Pandas/numpy:** For the array of coordinates
* **Geopy:** For a more accurate distance calculations using geodesics
* **Matplotlib:** For ploting, adding data, set labels etc.
* **Cartopy:** For mercator projection map, and shape reader.
* **Shapely** For block masking.


```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import cartopy
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
from shapely.geometry import MultiPolygon, Point

# States and Provinces
def states(ax):
    # Load states/provinces shapefile
    shapename = 'admin_1_states_provinces_lakes'
    states_shp = shpreader.natural_earth(resolution='110m', category='cultural', name=shapename)
    
    # Creating a mask for the heatmap
    states = list(shpreader.Reader(states_shp).geometries())

    # Some states have multipoligons within them causing problems
    all_polygons = []
    for geom in states:
        if geom.is_empty:
            continue
        if geom.geom_type == 'MultiPolygon':
            # Extract individual polygons
            all_polygons.extend(list(geom.geoms))
        elif geom.geom_type == 'Polygon':
            all_polygons.append(geom)
            
    # Combining all state geometries into a multipoligon for masking
    # Making it global so it looks clean
    global usa_geoms
    usa_geoms = MultiPolygon(all_polygons)
    
    for state in shpreader.Reader(states_shp).records():
    
        facecolor = 'none' #Transparent
        edgecolor = 'black'
    
        ax.add_geometries(
            [state.geometry],
            crs=ccrs.PlateCarree(),
            facecolor=facecolor,
            edgecolor=edgecolor,
            linewidth=0.4,
            zorder=2 # to the back
        )

# Basic US map with states and provinces
def usMap():
    # Init plot
    fig = plt.figure()
    
    # Setup LambertConformal projection map of USA
    ax = fig.add_subplot(1,1,1, projection=ccrs.LambertConformal())
    
    # USA coordinates range from -124.67 to -66.95 lon, 25.84 to 49.38 lat.
    ax.set_extent([-125, -66.5, 20, 50], crs=ccrs.Geodetic())

    # Adding States/Provinces
    states(ax)
    
    return fig, ax
```

```python
# Map
fig, ax = usMap()

# Ploting graph
plt.title('US States Map')
plt.show()
```

### **3. Pseudocode**

```python
"""
def calcDists(array:(lat, lon), custm:[customerCities: lat,lon,nShipments] -> sumDistArray
    This function will calculate the sum of all the distances from each customers
    and return an array of the same size

def twoMin(sumDistarray) -> coor1:(float, float) , coor2(float, float)
    This function will find the two lowest points and check if they are not touching.
    Will return two coordinates

def repart(coor1, coor2) -> custm1:[customerCities: lat,lon], custm2:[customerCities: lat,lon]
    This function will repart each customer to each coordinate with the minimum distance
    and return two lists

def arrayRes(coor:(float, float), res:int) -> array(lat, lon)
    This function will break down a coordinate or a range of coordinates into a smaller array 
    with a set resolution and returning an array

def main():

    array = arrayRes()
    wghtArr = calcDists(array)
    coor1, coor2 = twoMin(wghtArr)
    custmList1, custmList2 = repart(coor1, coor2)

    array1, array2 = arrayRes(coor1), arrayRes(coor2)
    wghtArr1, wghtArr2 = calcDists(array1, custmList1), calcDists(array2, custmList2)
    coor1, coor2 = min(wghtArr1), min(wghtArr2)

    plot(coor1, coor2)
"""
```

```python
import numpy as np
from geopy import distance
import time
```

```python
def coorsGrid(lon, lat, res):
    """
    Creates a grid of coordinates
    lon: [left, right] longitude
    lat: [south, north] lattitude
    res: Resolution (lower for more resolution)
    Returns a lon and lan mesh, and an array of zeros
    """
    lon_grid = np.arange(lon[0], lon[1], res)
    lat_grid = np.arange(lat[0], lat[1], res)

    lon_mesh, lat_mesh = np.meshgrid(lon_grid, lat_grid)

    newWeights = np.zeros_like(lon_mesh)

    return lon_mesh, lat_mesh, newWeights

    
    

def calcDists(lon_mesh, lat_mesh, newWeights, customers, byFlags=''):
    
    for i in range(lon_mesh.shape[0]):
        for j in range(lon_mesh.shape[1]):
            distSum = 0        
            for _, row in customers.iterrows():
                if byFlags and row['flags'] != byFlags:
                    continue
                dist = distance.distance(
                    (lat_mesh[i,j], lon_mesh[i,j]), 
                    (row['Latitude'], row['Longitude'])
                ).miles
                # Multiply distance by shipments for weight
                distSum += dist * row['SHIPMENTS']
            
            newWeights[i, j] = distSum

    return newWeights


    
        

def coorMin(lon_mesh, lat_mesh, weights, mode = 1):
    """
    Finds the minimum total weighted distance for two points
    Constraints: Each destination is assigned to one of the points
    returns two coordinates and their indexes in the weight map

    Pick two lowest-cost locations
    Ensure not the same and not too close
    Perform local search
    """
    min1, min2 = (0,0)
    minSumBuffer = weights.max()
    minCoorBuffer = (0,0)
    minDist = 100

    # Weight index
    wIndex1 = (0,0)
    wIndex2 = (0,0)

    # Find first point
    for i in range(lon_mesh.shape[0]):
        for j in range(lon_mesh.shape[1]):
            if weights[i,j] <= minSumBuffer:
                minSumBuffer = weights[i,j]
                minCoorBuffer = (lat_mesh[i,j], lon_mesh[i,j])
                wIndex1 = (i,j)

    min1 = minCoorBuffer
    minSumBuffer = weights.max()
    minCoorBuffer = (0,0)

    if mode == 2:
        return min1, wIndex1

    dist = distance.distance(
                    (lat_mesh[i,j], lon_mesh[i,j]), 
                    (min1[0], min1[1])
                ).miles

    # Find second point with a minimun distance of the first
    for i in range(lon_mesh.shape[0]):
        for j in range(lon_mesh.shape[1]):
            if weights[i,j] <= minSumBuffer:
                dist = distance.distance(
                    (lat_mesh[i,j], lon_mesh[i,j]), 
                    (min1[0], min1[1])
                ).miles
                if dist >= minDist:
                    minSumBuffer = weights[i,j]
                    minCoorBuffer = (lat_mesh[i,j], lon_mesh[i,j])
                    wIndex2 = (i,j)
                    
    min2 = minCoorBuffer

    return min1, min2, wIndex1, wIndex2


def repart(coors1, coors2, df):
    """
    Assign each customer to one of the two points
    returns a list
    """

    dist_to1 = 0
    dist_to2 = 0
    flags = []
    for _, row in df.iterrows():
        clientCoor = (row['Latitude'], row['Longitude'])
        dist_to1 = distance.distance(
            (coors1[0], coors1[1]), 
            (clientCoor[0], clientCoor[1])
            ).miles

        dist_to2 = distance.distance(
            (coors2[0], coors2[1]), 
            (clientCoor[0], clientCoor[1])
            ).miles

        if dist_to1 < dist_to2:
            flags.append('to1')
        else:
            flags.append('to2')

    return flags

start = time.time()
#Set USA range Array lon(-125, -66.5) lat(20, 50)
res = 0.2 # Resolution, corresponds to degrees
longitude = [-125, -66.5]
latitude = [20, 55]

# Create coordinates mesh
lon_mesh, lat_mesh, weights = coorsGrid(longitude, latitude, res)

# Populate weight mesh with values
weights = calcDists(lon_mesh, lat_mesh, weights, df)

# Obtain the two minum sum coordinates, and ensure they are not too close
# mode 2 returns just one coordinate
coor1, coor2, wIndex1, wIndex2 = coorMin(lon_mesh, lat_mesh, weights, mode = 1)

# Compares each distance to customer and designates their warehouse
flags = repart(coor1 , coor2, df)
#Append flags to dataframe
df['flags'] = flags

"""
Recalculate and refine
"""

# Recreate coordinates mesh at higher res
lon_mesh, lat_mesh, new_weights = coorsGrid(longitude, latitude, res)

# Recalculate for coordinate 1
# Populate weight mesh with values with flags now
weights1 = calcDists(lon_mesh, lat_mesh, np.zeros_like(weights), df, byFlags='to1')

# Obtain the minum sum coordinates
coor1_iter2, wIndex1 = coorMin(lon_mesh, lat_mesh, weights1, mode = 2)


# Recalculate for coordinate 2

# Populate weight mesh with values with flags now
weights2 = calcDists(lon_mesh, lat_mesh, np.zeros_like(weights), df, byFlags='to2')

# Obtain the minum sum coordinates
coor2_iter2, wIndex2 = coorMin(lon_mesh, lat_mesh, weights2, mode = 2)

end = time.time()
print(f'Elapsed runtime: {(end-start):.03f}s')

```

```python
# Ploting heatmap
# Plot US
fig, ax = usMap()

# Grid mask
lon_grid, lat_grid = np.meshgrid(lon_mesh[0], lat_mesh[:,0])
mask = np.zeros_like(lon_grid, dtype=bool)

for i in range(lon_grid.shape[0]):
    for j in range(lon_grid.shape[1]):
        point = Point(lon_grid[i,j], lat_grid[i,j])
        mask[i,j]=usa_geoms.contains(point)
weightsMasked = np.where(mask, weights, np.nan)

# Plot heatmap
heatmap = ax.pcolormesh(
    lon_mesh,
    lat_mesh,
    weightsMasked,
    transform=ccrs.PlateCarree(), #Important!
    cmap='viridis',
    shading='auto',
    zorder=0
)

# Colorbar
cbar = plt.colorbar(heatmap, ax=ax, label='Optimal Coverage Distance (mi)')
cbar.set_alpha(0.1)

# Fig size
fig.set_size_inches((12,6))
fig.set_dpi(100)

# Title and show
plt.title(f'Optimal Warehouse Locations Based on Distance (Res: {res})')
plt.show()
```

```python
# Ploting heatmap with selected zones, first pass
# Plot US
fig, ax = usMap()

# Grid mask
lon_grid, lat_grid = np.meshgrid(lon_mesh[0], lat_mesh[:,0])
mask = np.zeros_like(lon_grid, dtype=bool)

for i in range(lon_grid.shape[0]):
    for j in range(lon_grid.shape[1]):
        point = Point(lon_grid[i,j], lat_grid[i,j])
        mask[i,j]=usa_geoms.contains(point)

weightsMasked = np.where(mask, weights, np.nan)
# Plot heatmap
heatmap = ax.pcolormesh(
    lon_mesh,
    lat_mesh,
    weightsMasked,
    transform=ccrs.PlateCarree(), #Important!
    cmap='viridis',
    shading='auto',
    zorder=0
)

# Colorbar
cbar = plt.colorbar(heatmap, ax=ax, label='Optimal Coverage Distance (mi)')
cbar.set_alpha(0.1)

# Adding selected zones
xy = (float(coor1[1])-(res/2), float(coor1[0])-(res/2))
ax.add_patch(plt.Rectangle(
    xy,
    width=res,
    height= res,
    edgecolor='red',
    zorder=1,
    fill=False,
    transform=ccrs.PlateCarree()
))

xy = (float(coor2[1])-(res/2), float(coor2[0])-(res/2))
ax.add_patch(plt.Rectangle(
    xy,
    width=res,
    height= res,
    edgecolor='red',
    zorder=1,
    fill=False,
    transform=ccrs.PlateCarree()
))

# Fig size
fig.set_size_inches((12,6))
fig.set_dpi(100)

# Title and show
plt.title(f'Optimal Warehouse Locations Based on Distance (Res: {res})')
plt.show()
```

```python
# Ploting heatmap with selected zones, Second pass
# Plot US
fig, ax = usMap()

# Grid mask
lon_grid, lat_grid = np.meshgrid(lon_mesh[0], lat_mesh[:,0])
mask = np.zeros_like(lon_grid, dtype=bool)

for i in range(lon_grid.shape[0]):
    for j in range(lon_grid.shape[1]):
        point = Point(lon_grid[i,j], lat_grid[i,j])
        mask[i,j]=usa_geoms.contains(point)

# mesh with the two heatmaps
weights = np.minimum(weights1, weights2)
weightsMasked = np.where(mask, weights, np.nan)

# Plot heatmap
heatmap = ax.pcolormesh(
    lon_mesh,
    lat_mesh,
    weightsMasked,
    transform=ccrs.PlateCarree(), #Important!
    cmap='viridis',
    shading='auto',
    zorder=0
)

# Colorbar
cbar = plt.colorbar(heatmap, ax=ax, label='Optimal Coverage Distance (mi)')
cbar.set_alpha(0.1)

# Adding selected zones
# res = 0.5
xy = (float(coor1_iter2[1])-(res/2), float(coor1_iter2[0])-(res/2))
ax.add_patch(plt.Rectangle(
    xy,
    width=res,
    height= res,
    edgecolor='red',
    zorder=1,
    fill=False,
    transform=ccrs.PlateCarree()
))

xy = (float(coor2_iter2[1])-(res/2), float(coor2_iter2[0])-(res/2))
ax.add_patch(plt.Rectangle(
    xy,
    width=res,
    height= res,
    edgecolor='red',
    zorder=1,
    fill=False,
    transform=ccrs.PlateCarree()
))

# Fig size
fig.set_size_inches((12,6))
fig.set_dpi(100)

# Title and show
plt.title(f'Optimal Warehouse Locations Based on Distance (Res: {res})')
plt.show()
```

```python
# Comparing against
"""
The coordinates calculated should outperform randomly selected ones by being lower in average
"""
perf1 = 0
perf2 = 0

for test in range(1000):
    rand_lons = np.random.randint(0, weights.shape[0], 2,)
    rand_lats = np.random.randint(0, weights.shape[1], 2)
    
    perf1 += (weights[wIndex1[0],wIndex1[1]])/(weights[rand_lons[0],rand_lats[0]])
    perf2 += (weights[wIndex2[0],wIndex2[1]])/(weights[rand_lons[1],rand_lats[1]])

print(f'Performance 1: -{(1 - perf1/1000)*100:.2f}%')
print(f'Performance 2: -{(1 - perf2/1000)*100:.2f}%')
```
